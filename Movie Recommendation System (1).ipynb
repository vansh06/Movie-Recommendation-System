```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
d = pd.read_csv("u.data",sep="\t",names = 'user_id item_id rating timestamp'.split())

d
```

```python
col_name = ['item_id','title']+[str(i) for i in range(22)]
f= pd.read_csv("u.item",sep='|',names = col_name)
```

```python
f
```

```python
f.columns
```

```python
f1=f[['item_id','title']]
f1
```

```python
e=pd.merge(d,f1,on = 'item_id')
e
```

```python
f=e.groupby("title")
```

```python
f["rating"].mean().sort_values(ascending=False)
```

```python
rating= pd.DataFrame(f["rating"].mean().sort_values(ascending=False))
rating
```

```python
rating["r_count"] = e["title"].value_counts()
rating
```

```python
rating.sort_values('r_count', ascending = False)
```

```python
pivot=pd.pivot_table(e,values="rating",index="user_id",columns="title")
pivot
```

```python
df_matrix = pivot[input()]
```

```python
df_matrix
```

```python
corr = pivot.corrwith(df_matrix)
corr
```

```python
df_corr= pd.DataFrame(corr, columns=['correlation'])
```

```python
df_corr.dropna(inplace=True)
```

```python
df_corr
```

```python
df_corr.sort_values(by="correlation",ascending=False)
```

```python
df_corr = df_corr.join(rating["r_count"])
```

```python
df_corr
```

```python
df_corr[df_corr["r_count"]>200].sort_values(by="correlation",ascending=False).head(30)
```

```python
df_matrix = pivot[input()]
corr = pivot.corrwith(df_matrix)
df_corr= pd.DataFrame(corr, columns=['correlation'])
df_corr.dropna(inplace=True)
df_corr.sort_values(by="correlation",ascending=False)
df_corr = df_corr.join(rating["r_count"])
df_corr[df_corr["r_count"]>200].sort_values(by="correlation",ascending=False).head(30)
```
